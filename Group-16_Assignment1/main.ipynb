Creating a custom tag by POS tagging

```python
matcher = Matcher(nlp.vocab)

ingredient_patterns = [
    [{"POS": "NOUN", "OP": "+"}],  # e.g., "chicken broth"
    [{"POS": "ADJ"}, {"POS": "NOUN", "OP": "+"}],  # e.g., "minced garlic"
    [{"POS": "PROPN"}, {"POS": "NOUN", "OP": "+"}],  # e.g., "Dijon mustard"
]

# Add patterns to the matcher
for i, pattern in enumerate(ingredient_patterns):
    matcher.add(f"Ingredient_{i}", [pattern])

# Custom component to identify ingredients
@Language.component("ingredient_component")
def ingredient_component(doc):
    matches = matcher(doc)
    spans = [Span(doc, start, end, label="INGREDIENT") for _, start, end in matches]
    spans = spacy.util.filter_spans(spans) 
    doc.ents = spans 
    return doc

nlp.add_pipe("ingredient_component", last=True)
```

# 1b. Applying NER

```python
# List of common non-ingredient words
non_ingredient_words = {"cups", "cup", "tablespoons", "tablespoon", "teaspoons", "teaspoon", "ounce", "ounces", "pinch", "pound", "pounds", "stirring", "degrees", "room temperature"}

def extract_ingredients(text):
    doc = nlp(text)
    ingredients = []
    
    for ent in doc.ents:
        if ent.label_ == "INGREDIENT":
            ingredient_tokens = [token for token in ent if token.pos_ in ["NOUN", "PROPN", "ADJ"]]
            # Filter out non-ingredient words only if they are the first token
            if ingredient_tokens and ingredient_tokens[0].text.lower() not in non_ingredient_words:
                ingredient_name = " ".join([token.text for token in ingredient_tokens])
                if ingredient_name:  # Ensure non-empty strings
                    ingredients.append(ingredient_name)
    
    return ingredients

df['ingredient_names'] = df['recipe_Ingredients'].apply(lambda x: extract_ingredients(str(x)))

print(df['ingredient_names'].head())
```

# 1c. Recipe - Ingredient pair saved

```python
recipe_ingredient_pairs = []

for index, row in df.iterrows():
    recipe_id = row['recipe_id']
    for ingredient in row['ingredient_names']:
        recipe_ingredient_pairs.append((recipe_id, ingredient))

# Save the pairs to a .txt file
with open('recipe_ingredient_pairs.txt', 'w', encoding="utf-8") as file:
    for pair in recipe_ingredient_pairs:
        file.write(f"{pair[0]} - {pair[1]}\n")

print("Saved")
```

```python
import random

unique_recipe_ids = list(set([pair[0] for pair in recipe_ingredient_pairs]))
random_recipe_ids = random.sample(unique_recipe_ids, 100)
random_recipe_ingredient_pairs = [pair for pair in recipe_ingredient_pairs if pair[0] in random_recipe_ids]

with open('random100_recipe_ingredient_pairs.txt', 'w', encoding="utf-8") as file:
    for pair in random_recipe_ingredient_pairs:
        file.write(f"{pair[0]} - {pair[1]}\n")

print("Saved 100 random recipes")
```

# 2c

```python
recipe_sizes = np.array(recipe_sizes)
hist_data, bin_edges = np.histogram(recipe_sizes, bins=20, density=True)
cdf = np.cumsum(hist_data * np.diff(bin_edges))
cdf /= cdf[-1]

plt.figure(figsize=(10, 6))
plt.plot(bin_edges[1:], cdf, color='skyblue', marker='o', linestyle='-')

plt.title("Cumulative Recipe Size Distribution")
plt.xlabel("Recipe Size")
plt.ylabel("Cumulative Probability")

plt.show()
```
